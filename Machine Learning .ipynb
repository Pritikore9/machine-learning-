# Q1: Take 5 students and their marks 3 subject and calculate the total & average of each student.

```python
def school(n):
    name=[]
    for x in range(n):
        nm=input("Enter the name: ")
        name.append(nm)
    return name
def marks():
    details=[]
    marks=[]
    name=school(5)
    print("marks function: ")
    for x in name:
        print("my name is ",x)
        details.append(x)
        details.append("My marks")
        for x1 in range(3):
            m=int(input("Enter the marks: "))
            marks.append(m)
            details.append(m)
        details.append("Total MArks: ")
        total=sum(marks)
        details.append(total)
        details.append("Average: ")
        avg=total/len(marks)
        details.append(avg)
        print("My marks are: ", details)
        marks.clear()
marks()
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
data["Class"]=l.fit_transform(data["Class"])
```

```python
data
```

```python
data["Class"].value_counts()
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

## Imbalancing 

```python
from imblearn.over_sampling import RandomOverSampler
r=RandomOverSampler()
x_data,y_data=r.fit_resample(x,y)
```

```python
from collections import Counter
print(Counter(y_data))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.3,random_state=51)
```

```python
print(Counter(y_train))
```

```python
print(len(x_train))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
pred=knn.predict(x_test)
pred
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.metrics import accuracy_score
nb=GaussianNB()
nb.fit(x_train,y_train)
nb_pred=nb.predict(x_test)
nb_acc=accuracy_score(nb_pred,y_test)*100
print("Accuracy score of naive bayes: ",nb_acc)
```

```python
print((x_train).shape)
print((x_test).shape)
print((y_train).shape)
print((y_test).shape)
```

```python
from sklearn.model_selection import StratifiedKFold
skf=StratifiedKFold()
skf.get_n_splits(x_data,y_data)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
```

```python
for train_data,test_data in skf.split(x_data,y_data):
    x_train,x_test=x_data[train_data],x_data[test_data]
    y_train,y_test=y_data[train_data],y_data[test_data]
    
    scores=cross_val_score(nb,x_train,y_train,cv=skf)
    y_pred=cross_val_predict(nb,x_test,y_test)
    
    print("Checking the balance: ",Counter(y_test))
    print(scores)
    print(y_pred)
    print(np.mean(scores)*100)
```
